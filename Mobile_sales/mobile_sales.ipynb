# 8. Which are the top 10 models by avg sales


```python
avg_sales_by_model= df.groupby('model')['sales'].mean()
```

```python
top_10_model= avg_sales_by_model.sort_values(ascending=False).head(10)
```

```python
top_10_model
```

```python
top_10_model.plot(kind='bar')
plt.xlabel('Model')
plt.ylabel('Sales in corer')
for i, value in  enumerate(top_10_model):
    plt.text(i,value,str(round(value,2)), ha='center', va='bottom')
    
plt.tight_layout()
plt.show()   
```
